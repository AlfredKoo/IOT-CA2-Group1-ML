```python
def get_avg(featurearray):
    buffer = []
    for row in featurearray:
        try:
            value = mean(row)
        except:
            value = 0
        # Can choose if you want to round off or not
        buffer.append(round(value))
    return buffer
```

```python
def get_mode(featurearray):
    buffer = []
    for devicerow in featurearray:
        try:
            value = mode(devicerow)
        except:
            value = 0
        buffer.append(value)  # Can choose if you want to round off or not
    return buffer
```

```python
def get_max(featurearray):
    buffer = []
    for devicerow in featurearray:
        try:
            value = max(devicerow)
        except:
            value = 0
        buffer.append(value)  # Can choose if you want to round off or not
    return buffer
```

### Establish targets X and Y

```python
df.head()
```

```python
# Making X your variable and Y your target to predict
target = 'label'
features = ['Accuracy', 'Bearing', 'acceleration_x', 'acceleration_y', 'acceleration_z','gyro_x','gyro_y','gyro_z','Speed'] #For the everything else
print(target)
print(features)
```

```python
x = df.loc[:, features] #Your input. Take all row and all columns in features
y = df.loc[:, target] #Take  all row and only survived column
print(x.head(5))
print(y.head(5))
print(type(x))
print(type(y))
```

```python
#Split data set into training and testing
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25) #You would want to reshuffle your data
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

### Normalize features through scaling

```python
#Scaling ensures no variable have an advantage over another. Use same scaler for x_train and x_test
x_scaler = StandardScaler() 
z_train = x_scaler.fit_transform(x_train)
```

```python
z_train = pd.DataFrame(data = z_train, columns = features ) #input data is z_train, specified columns
```

### Model Selection(SGDClassifier)

```python
model = SGDClassifier() #Non-optimised
model.fit(x_train,y_train) #Training scaled x to y
model_nameSGD='SGDRegressor'
y_hatSGD = model.predict(z_test)
```

```python
r2scoreSGD=format(r2_score(y_test,y_hatSGD),'.2f')
mseSGD=format(mean_squared_error(y_test,y_hatSGD),'.2f')
rtrsmSGD= float(format(model.score(z_train, y_train),'.3f')) #R-squared training is used to correlate how close the prediction is to the values
accuracySGD=str(accuracy_score(y_test,y_hatSGD))
print('Accuracy Score : ' + str(accuracy_score(y_test,y_hatSGD)))
print('R2:',"{:.2f}".format(r2_score(y_test,y_hatSGD))) #Range of error #If 1 is perfect score
print('MSE:',"{:.2f}".format(mean_squared_error(y_test,y_hatSGD)))
```
